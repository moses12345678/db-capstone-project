```python
import mysql.connector
from mysql.connector import Error
from mysql.connector.pooling import MySQLConnectionPool

try:
    # Step 1: Define your database configurations as a Python dictionary object called dbconfig.
    dbconfig = {
        "database": "little_lemon_db",
        "user": "root",
        "password": "",
        "host": "localhost",  # Change to your database host if not running locally.
    }

    # Step 2: Create a connection pool named pool_b with two connections.
    pool_b = MySQLConnectionPool(pool_name="pool_b", pool_size=2, **dbconfig)

    # Step 3: Get a connection from the pool.
    try:
        connection = pool_b.get_connection()

        # Step 4: Create a buffered cursor.
        cursor = connection.cursor(buffered=True)

        # Step 5: Combine the data from the Bookings and the Employee tables and sort the retrieved records.
        query = """
        SELECT B.BookingSlot, CONCAT(GuestFirstName, ' ', GuestLastName) AS GuestName,
               CONCAT('Assigned to: ', Name, ' [', Role, ']') AS EmployeeInfo
        FROM Bookings B
        LEFT JOIN Employees E ON B.EmployeeID = E.EmployeeID
        ORDER BY B.BookingSlot ASC
        LIMIT 3
        """

        cursor.execute(query)

        # Step 6: Display the information of the first three guests.
        print("Upcoming Bookings:")
        for row in cursor:
            booking_slot, guest_name, employee_info = row
            print("[{}] [{}] [{}]".format(booking_slot, guest_name, employee_info))

    except mysql.connector.errors.PoolError as e:
        print("PoolError:", e)

    finally:
        # Step 7: Return the connection back to the pool.
        if connection is not None:
            connection.close()

except Error as e:
    print("Error while connecting to MySQL:", e)
```
